```python
def calcular_veiculos():
    # Solicitar o número de pessoas
    num_pessoas = int(input("Quantas pessoas vão viajar? "))

    # Solicitar a capacidade do ônibus
    capacidade_onibus = int(input("Qual a capacidade do ônibus? "))

    # Solicitar a capacidade de cada van
    capacidade_van = int(input("Qual a capacidade de cada van? "))

    # Verificar se é possível alugar um único ônibus
    if num_pessoas <= capacidade_onibus:
        print(f"Você pode alugar 1 ônibus para acomodar {num_pessoas} pessoas.")
    else:
        # Calcular a quantidade de ônibus necessários
        num_onibus = num_pessoas // capacidade_onibus
        if num_pessoas % capacidade_onibus != 0:
            num_onibus += 1
        print(f"Você precisa de {num_onibus} ônibus para acomodar {num_pessoas} pessoas.")

    # Verificar se é possível alugar vans
    num_vans = num_pessoas // capacidade_van
    if num_pessoas % capacidade_van != 0:
        num_vans += 1

    print(f"Você precisa de {num_vans} vans para acomodar {num_pessoas} pessoas.")

# Rodar a função
calcular_veiculos()
3
```
